```python
# ── Cell 1: 导入库 & 读取文本文件 ──
import re
import pandas as pd

# 请根据实际路径修改
file_path = "Processed_Result_BD.txt"

with open(file_path, 'r', encoding='utf-8') as f:
    lines = [line.strip() for line in f if line.strip()]
```

```python
# ── Cell 2: 解析行，提取字段 ──
records = []
# 更通用的正则：直接抓 host 整体，再在后面拆字母和数字
pattern = re.compile(
    r'^(?P<ts>\S+)\s+'                  # 时间戳
    r'(?P<host>[A-Za-z]+\d+)\s+'        # host e.g. A05, D12
    r'round\s+\d+:\s+'
    r'Phi_CSI=(?P<phi_csi>[-\d\.]+),.*?'# Phi_CSI
    r'avg_ampl=\[\s*(?P<avg0>[-\d\.]+)' # avg_ampl[0]
)

for line in lines:
    m = pattern.match(line)
    if not m:
        # 如果有不匹配的行，可以打印出来调试
        # print("跳过：", line)
        continue

    host = m.group('host')
    # 拆分字母和数字
    lm = re.match(r'^([A-Za-z]+)(\d+)$', host)
    letters, number = lm.group(1), int(lm.group(2))

    records.append({
        'host_letters': letters,
        'host_number': number,
        'host': host,
        'Phi_CSI': float(m.group('phi_csi')),
        'avg_ampl_0': float(m.group('avg0')),
    })

df = pd.DataFrame(records)
print(f"共解析到 {len(df)} 条记录。")
```

```python
# ── Cell 3: 排序 & 显示结果 ──
df_sorted = (
    df
    .sort_values(by=['host_letters','host_number'], ignore_index=True)
    [['host','Phi_CSI','avg_ampl_0']]
)
df_sorted
```
